```python
import numpy as np
```

```python
city_list = ['A','B','C','D','E','F']
def plot_cities(city_list):
    city_coords = {}         #will store the coordinates of each city
    city_n = len(city_list)
    grid_size = 100*city_n
    for i in range(city_n):          #Applying a random function to random assign the coordinates to the city.
        city = city_list[i]
        x = np.random.randint(grid_size)
        y = np.random.randint(grid_size)
        city_coords[city] = [x,y]             #storing the coords in a dictionary. [0] is x, [1] is y.
    
    return city_coords

city_coords = plot_cities(city_list)
city_coords
```

```python


def find_dist(city_coords):
    city_dist = {}
    for city1 in city_coords:
        sub_dict = {}
        for city2 in city_coords:
            if (city1 == city2):
                sub_dict[city1] = 0               #0 distance implies same city , leads to no change in distance, so it will never be used. Maybe find a better way of doing this.
                continue
            x1, x2, y1, y2 = city_coords[city1][0], city_coords[city2][0], city_coords[city1][1], city_coords[city2][1]
            sub_dict[city2] = ((x1-x2)**2 + (y1-y2)**2)**0.5             #getting the distance. 
        
        city_dist[city1] = sub_dict
    return city_dist

city_dist = find_dist(city_coords)
city_dist
```
